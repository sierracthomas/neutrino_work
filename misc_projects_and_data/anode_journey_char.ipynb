```python
def myplot(ax, x, light, dark, mytitle):
    ax.scatter(x, dark, label = "dark current")
    ax.scatter(x[:len(light)], light, label = "light current")
    ax.legend()
    ax.set_yscale("log")
    ax.title.set_text(mytitle)
    plt.xlabel("volts")
    plt.ylabel(r"$\mu$A")
    
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
    
    
    
myplot(ax[0], x480, y1480, y480, "480 nm, SiPM 403")

myplot(ax[1], vout, y_light, y_dark, "280 nm, SiPM 402")




plt.legend()
plt.show()
```
